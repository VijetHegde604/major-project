1. Bar Plot – Per-Class mAP (YOLOv8)
Purpose: Visualize detection performance across classes from YOLOv8.

```python
import matplotlib.pyplot as plt

classes = ['Charminar', 'Gateway of India', 'Mysore Palace', 'Qutub Minar', 'Taj Mahal']
mAP_values = [0.523, 0.705, 0.404, 0.492, 0.493]

plt.figure(figsize=(8, 5))
plt.bar(classes, mAP_values, color='skyblue')
plt.title("YOLOv8 Per-Class mAP@0.5:0.95")
plt.ylabel("mAP Score")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

2. Confusion Matrix – Hybrid Model Classification
Purpose: Show how well the hybrid classifier predicted true labels.

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load data
pred_df = pd.read_csv("/home/vijeth/major-project/AI/hybrid_model/results/predictions.csv")
gt_df = pd.read_csv("/home/vijeth/major-project/AI/hybrid_model/results/ground_truth.csv")

# Merge on image_name
merged = pd.merge(gt_df, pred_df, on="image_name")

# Extract true and predicted labels
y_true = merged["true_label"]
y_pred = merged["predicted_label"]
labels = sorted(y_true.unique())

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Confusion Matrix – Hybrid Classification")
plt.tight_layout()
plt.show()
```
